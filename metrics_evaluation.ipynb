```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```python
file_path = './target/evaluation.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
print('J48 model metrics evaluation:\n')
print('Accuracy Mean: ', data[data['Model'] == 'J48']['Accuracy'].mean()) 
print('Accuracy Median: ', data[data['Model'] == 'J48']['Accuracy'].median())

print('Precision Mean: ', data[data['Model'] == 'J48']['Precision'].mean()) 
print('Precision Median: ', data[data['Model'] == 'J48']['Precision'].median())

print('Recall Mean: ', data[data['Model'] == 'J48']['Recall'].mean()) 
print('Recall Median: ', data[data['Model'] == 'J48']['Recall'].median())

print('F1-Score Mean: ', data[data['Model'] == 'J48']['F1 Score'].mean()) 
print('F1-Score Median: ', data[data['Model'] == 'J48']['F1 Score'].median())

print('True Pos Mean: ', data[data['Model'] == 'J48']['True Pos'].mean()) 
print('True Pos Median: ', data[data['Model'] == 'J48']['True Pos'].median())

print('True Neg Mean: ', data[data['Model'] == 'J48']['True Neg'].mean()) 
print('True Neg Median: ', data[data['Model'] == 'J48']['True Neg'].median())

print('False Pos Mean: ', data[data['Model'] == 'J48']['False Pos'].mean()) 
print('False Pos Median: ', data[data['Model'] == 'J48']['False Pos'].median())

print('False Neg Mean: ', data[data['Model'] == 'J48']['False Neg'].mean()) 
print('False Neg Median: ', data[data['Model'] == 'J48']['False Neg'].median())

plt.scatter(data[data['Model'] == 'J48']['Accuracy'], data[data['Model'] == 'J48']['F1 Score'])
plt.xlabel('Accuracy')
plt.ylabel('F1 Score')
```
