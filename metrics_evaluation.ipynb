```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pyplot as plt
```

```python
file_path = './target/evaluation.csv'
data = pd.read_csv(file_path)
data.head()

lburst_path = './uc10_lburst_metrics'
sburst_path = './uc10_sburst_metrics'
linterval_path = './uc10_linterval_metrics'
minterval_path = './uc10_mininterval_metrics'
sinterval_path = './uc10_sinterval_metrics'

# create these directories if they do not exist
import os
os.makedirs(lburst_path, exist_ok=True)
os.makedirs(sburst_path, exist_ok=True)
os.makedirs(linterval_path, exist_ok=True)
os.makedirs(minterval_path, exist_ok=True)
os.makedirs(sinterval_path, exist_ok=True)
```

```python
if (data[data['Model'] == 'J48']['Accuracy'].mean() == data[data['Model'] == 'RandomForest']['Accuracy'].mean() and (data[data['Model'] == 'J48']['F1 Score'].mean() == data[data['Model'] == 'RandomForest']['F1 Score'].mean()) and (data[data['Model'] == 'J48']['Precision'].mean() == data[data['Model'] == 'RandomForest']['Precision'].mean())):
    print("Equal results Random Forest and J48")

if (data[data['Model'] == 'J48']['Accuracy'].mean() == data[data['Model'] == 'NaiveBayes']['Accuracy'].mean() and (data[data['Model'] == 'J48']['F1 Score'].mean() == data[data['Model'] == 'NaiveBayes']['F1 Score'].mean()) and (data[data['Model'] == 'J48']['Precision'].mean() == data[data['Model'] == 'NaiveBayes']['Precision'].mean())):
    print("Equal results Naive Bayes and J48")

if (data[data['Model'] == 'J48']['Accuracy'].mean() == data[data['Model'] == 'REPTree']['Accuracy'].mean() and (data[data['Model'] == 'J48']['F1 Score'].mean() == data[data['Model'] == 'REPTree']['F1 Score'].mean()) and (data[data['Model'] == 'J48']['Precision'].mean() == data[data['Model'] == 'REPTree']['Precision'].mean())):
    print("Equal results REPTree and J48")
```
